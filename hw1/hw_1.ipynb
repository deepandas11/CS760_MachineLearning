```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python

p_values = [1, 2, np.inf]
xx, yy = np.meshgrid(np.linspace(-3, 3, num=101), np.linspace(-3, 3, num=101))
fig, axes = plt.subplots(ncols=3,
                     nrows=1, figsize=(20, 5))

# fig.suptitle('L- based Norms Region Visualizations', size=10)

for p, ax in zip(p_values, axes.flat):
    if p == 0:
        zz = (xx != 0).astype(int) + (yy != 0).astype(int)
        ax.imshow(zz, cmap='bwr', extent=(xx.min(),xx.max(),yy.min(),yy.max()), aspect="auto")
    else:
        if np.isinf(p):
            zz = np.maximum(np.abs(xx),np.abs(yy))
            st = 'L-inf Norm'
        else:
            zz = ((np.abs((xx))**p) + (np.abs((yy))**p))**(1./p)
            st = 'L-'+str(p)+' Norm'

        ax.contourf(xx, yy, zz, 100, cmap='inferno')
        ax.contour(xx,yy,zz, [1], colors='red', linewidths = 2) 
        ax.set_title(st, size=30)


plt.show()
```

```python
mean = [0,0]
cov = [[1, 0], [0, 1]]

x, y = np.random.multivariate_normal(mean, cov, 100).T

plt.grid()
plt.scatter(x, y)
plt.ylim(-10, 10)
plt.xlim(-10,10)

plt.xlabel("X coordinate")
plt.ylabel("Y coordinate")


plt.title("Zero Mean, Identity Covariance Matrix", size=20)
```
